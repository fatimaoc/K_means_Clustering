### Choosing the Appropriate Number of Clusters 

#### The elbow method

```python
kmeans_kwargs = {
    "init": "random",
    "n_init": 10,
    "max_iter": 300,
    "random_state": 42,   
}
#liste avec les valeurs SSE pour chaque k
sse = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, **kmeans_kwargs)# ** : opérateur de décompression , permet de parcourir un dict
    kmeans.fit(scaled_features)
    sse.append(kmeans.inertia_)
    
    
```

#### Visualisation de la courbe du SSE en fonction du nombres de clusters 

```python
plt.style.use("fivethirtyeight")
plt.plot(range(1, 11), sse)
plt.xticks(range(1, 11))
plt.xlabel("Number of Clusters")
plt.ylabel("SSE")
plt.show()
```

On peut voir que l'elbow point se situe sur 3 pour le nombre de clusters

#### Identification of the elbow point by programming 

```python
kl = KneeLocator(
    range(1, 11), sse, curve="convex", direction="decreasing")
kl.elbow
```

Cela confirme que l'elbow point visible sur la courbe se situe bien  sur 3

#### The silhouette coefficient

```python
silhouette_coefficients = []
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, **kmeans_kwargs)
    kmeans.fit(scaled_features)
    score = silhouette_score(scaled_features, kmeans.labels_)
    silhouette_coefficients.append(score)
    
```

#### Visualisation de la courbe du coefficient de silhouette en fonction du nombres de clusters  

Le coefficient de silhouette est une mesure de cohésion et de la séparation des groupes.
Les valeurs sont comprises entre -1 et 1.Les valeurs les plus élevés indiquent que les échantillons sont plus proches de leurs groupes.

```python
plt.plot(range(2, 11), silhouette_coefficients)
plt.xticks(range(2, 11))
plt.xlabel("Number of Clusters")
plt.ylabel("Silhouette Coefficient")
plt.show()
```

Selon la courbe ci-dessus, le meilleur choix du nombre de cluster est 3 .

#### Choix du nombres Clusters 

Selon les deux méthodes utilisées pour choisir le nombre optimal de clusters est 3

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
from sklearn.metrics import adjusted_rand_score
```

####  Use make_moons() to generate synthetic data in the shape of crescents 

```python
features, true_labels = make_moons(
    n_samples=250, noise=0.05, random_state=42
)
scaled_features = scaler.fit_transform(features)
```

```python
kmeans = KMeans(n_clusters=2)
dbscan = DBSCAN(eps=0.3)
kmeans.fit(scaled_features)
dbscan.fit(scaled_features)
kmeans_silhouette = silhouette_score(
    scaled_features, kmeans.labels_
).round(2)
dbscan_silhouette = silhouette_score(
   scaled_features, dbscan.labels_
).round (2)
```

#### Comparison of DBSCAN and k-means clustering results using ARI as a performance metric

```python
ari_dbscan = adjusted_rand_score(true_labels, dbscan.labels_)

round(ari_dbscan, 2)
```

Selon cette le métrique ARI, DBSCAN est le meilleur choix .

### Define our evaluation benchmark 

Compare different initialization methods for KMeans:
*create a pipeline which will scale the data using a StandardScaler*;
*train and time the pipeline fitting*;
*measure the performance of the clustering obtained via different metrics*.



```python
from time import time
from sklearn import metrics
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

def bench_k_means(kmeans, name, data, labels, use_stadard_scaler):
    """Benchmark to evaluate the KMeans initialization methods.

    Parameters
    ----------
    kmeans : KMeans instance
        A :class:`~sklearn.cluster.KMeans` instance with the initialization
        already set.
    name : str
        Name given to the strategy. It will be used to show the results in a
        table.
    data : ndarray of shape (n_samples, n_features)
        The data to cluster.
    labels : ndarray of shape (n_samples,)
        The labels used to compute the clustering metrics which requires some
        supervision.
    """
    t0 = time()
    if use_stadard_scaler:
        estimator = make_pipeline(StandardScaler(), kmeans).fit(data)
    else:
        estimator = make_pipeline(kmeans).fit(data)
    fit_time = time() - t0
    results = [name, fit_time, estimator[-1].inertia_]

    # Define the metrics which require only the true labels and estimator
    # labels
    clustering_metrics = [
        metrics.homogeneity_score,
        metrics.completeness_score,
        metrics.v_measure_score,
        metrics.adjusted_rand_score,
        metrics.adjusted_mutual_info_score,
        metrics.normalized_mutual_info_score,
    ]
    results += [m(labels, estimator[-1].labels_) for m in clustering_metrics]

    # The silhouette score requires the full dataset
    results += [
        metrics.silhouette_score(data, estimator[-1].labels_,
                                 metric="euclidean", sample_size=300,)
    ]

    # Show the results
    formatter_result = ("{:9s}\t{:.3f}s\t{:.0f}\t{:.3f}\t{:.3f}"
                        "\t{:.3f}\t{:.3f}\t{:.3f}\t{:.3f}\t{:.3f}")
    print(formatter_result.format(*results))
```

### Run the benchmark 

```python
use_st_sc = True
for i in range(2):
    print("standard scaler :", use_st_sc)
    print(82 * '_')
    print('init\t\ttime\tinertia\thomo\tcompl\tv-meas\tARI\tAMI\tNMI\tsilhouette')

    ## Random init
    kmeans = KMeans(init="random", n_clusters=10, n_init=4, random_state=0)
    bench_k_means(kmeans=kmeans, name="random", data=digits.data, labels=digits.target, use_stadard_scaler=use_st_sc)

    ## k-means++ init
    kmeans = KMeans(init="k-means++", n_clusters=10, n_init=4,
                    random_state=0)
    bench_k_means(kmeans=kmeans, name="k-means++", data=digits.data, labels=digits.target, use_stadard_scaler=use_st_sc)

    ## PCA components init
    pca = PCA(n_components=10).fit(digits.data)
    kmeans = KMeans(init=pca.components_, n_clusters=10, n_init=1)
    bench_k_means(kmeans=kmeans, name="PCA-based", data=digits.data, labels=digits.target, use_stadard_scaler=use_st_sc)

    print(82 * '_')
    print('')
    print('')
    use_st_sc = False
    
print('homo : homogeneity_score, compl : completeness_score, v-meas : v-measurment, ARI : adjusted_rand_score, AMI : adjusted_mutual_info_score, NMI : normalized_mutual_info_score')
```

Selon les résultats du benchmark , on peut en déduire que l'utilisation du standardScaler sur notre jeu de donnée diminue l'efficacité des performances de nos algorithmes.

### Visualize the results on PCA-reduced data
